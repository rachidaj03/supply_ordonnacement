```python
from ortools.sat.python import cp_model
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd

quantites = [30, 30, 30, 30, 30, 30, 30, 10, 10, 10, 10, 10, 10]

def ordonnancement_production(donnees):
    # Initialisation du modèle
    model = cp_model.CpModel()
    
    produits = set([d["Pièce"] for d in donnees])
    machines = set([d["Machine"] for d in donnees])
    phases = {p: [d for d in donnees if d["Pièce"] == p] for p in produits}
    
    # Variables
    start_times = {}
    end_times = {}
    all_tasks = []

    for d in donnees:
        task_id = (d["Pièce"], d["Phase"], d["Machine"])
        start_var = model.NewIntVar(0, 86400, f"start_{task_id}")
        end_var = model.NewIntVar(0, 86400, f"end_{task_id}")
        duration = int(d["T Série (H)"] * 3600 + d["T Unitaire (H)"] * d["Quantité"] * 3600)
        all_tasks.append((start_var, end_var, duration, d["Machine"]))
        start_times[task_id] = start_var
        end_times[task_id] = end_var

    # Contraintes : Ordre des phases pour chaque produit
    for p, tasks in phases.items():
        for i in range(len(tasks) - 1):
            current_phase = (tasks[i]["Pièce"], tasks[i]["Phase"], tasks[i]["Machine"])
            next_phase = (tasks[i + 1]["Pièce"], tasks[i + 1]["Phase"], tasks[i + 1]["Machine"])
            model.Add(end_times[current_phase] <= start_times[next_phase])

    # Contraintes : Une machine ne peut gérer qu'une tâche à la fois
    machine_tasks = {m: [] for m in machines}
    for task in all_tasks:
        start_var, end_var, duration, machine = task
        
        # Créer un IntervalVar pour chaque tâche
        interval_var = model.NewIntervalVar(start_var, duration, end_var, f"interval_{machine}")
        machine_tasks[machine].append(interval_var)
    
    for machine, intervals in machine_tasks.items():
        model.AddNoOverlap(intervals)

    # Objectif : Minimiser le makespan (fin la plus tardive)
    makespan = model.NewIntVar(0, 86400, "makespan")
    model.AddMaxEquality(makespan, [end for _, end, _, _ in all_tasks])
    model.Minimize(makespan)

    # Résolution
    solver = cp_model.CpSolver()
    status = solver.Solve(model)

    # Affichage des résultats
    if status == cp_model.OPTIMAL:
        
        print(f"Ordonnancement optimal trouvé avec un makespan de {solver.ObjectiveValue() / 3600} heures")
        for d in donnees:
            task_id = (d["Pièce"], d["Phase"], d["Machine"])
            print(f"Tâche {task_id}: Début = {solver.Value(start_times[task_id]) / 3600} h, "
                  f"Fin = {solver.Value(end_times[task_id]) / 3600} h")
    else:
        print("Pas de solution optimale trouvée.")
    

# Exemple de données structurées
donnees = pd.read_csv('/Users/rachidaitjalloul/Desktop/ÉTUDES/S7/SPL/ordonnancement_donnees.csv')
donnees = donnees.to_dict(orient="records")

prod = list(set(d["Pièce"] for d in donnees))  # Obtenir la liste unique des produits
prod.sort(key=lambda x: int(x[1:]))
quantites_dict = {prod[i]: quantites[i] for i in range(len(prod))}
for d in donnees:
    d["Quantité"] = quantites_dict[d["Pièce"]]

# Appeler la fonction
ordonnancement_production(donnees)
```
